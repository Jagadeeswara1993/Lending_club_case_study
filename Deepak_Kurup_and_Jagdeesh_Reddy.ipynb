### Checkpoint 2

We can further clean some more columns by dropping those columns which have only one unique value in it as it might not help us much with any calculation

```python
loan_data.nunique().sort_values()
```

```python
loan_data.drop(['tax_liens','delinq_amnt','chargeoff_within_12_mths','acc_now_delinq','application_type','policy_code','collections_12_mths_ex_med','initial_list_status','pymnt_plan'],axis=1,inplace=True)
```

```python
loan_data.shape
```

lets drop the some of the columns ,As we are only interested only in loan application details only.

```python
loan_data.drop(['emp_title', 'title','member_id', 'url','total_rec_int', 'total_rec_prncp', 'total_rec_late_fee', 'last_credit_pull_d', 'recoveries', 'collection_recovery_fee', 'last_pymnt_d'],axis=1,inplace=True)
loan_data.drop(['out_prncp_inv','out_prncp'],axis=1,inplace=True)
```

```python
loan_data.shape
```

```python
null_perc1=round(100*(loan_data.isnull().sum()/len(loan_data.index)),2)
```

```python
null_perc1[null_perc1>0]
```

removing the null valued rows of above 3 columns

```python
loan_data=loan_data[~loan_data.emp_length.isnull()]
loan_data=loan_data[~loan_data.revol_util.isnull()]
loan_data=loan_data[~loan_data.pub_rec_bankruptcies.isnull()]
```

```python
null_perc1=round(100*(loan_data.isnull().sum()/len(loan_data.index)),2)
```

```python
null_perc1[null_perc1>0]
```

```python
loan_data.isnull().sum()
```

```python
loan_data.shape
```

### Checkpoint 3

Further cleaning data for our analysis 
* Loan status has 3 distinct values - fully paid, charged off and current.So,we remove the rows which are current.
* Stripping extra space from terms
* Cleaning numeric fields like int_rate and revol_util
* Type conversion after clearing int_rate and revol_util

```python
loan_data=loan_data[loan_data.loan_status!='Current']
```

```python
loan_data['term'] = loan_data.term.str.strip()
loan_data.term.unique()
```

```python
loan_data.shape
```

```python
loan_data.dtypes
```

int_rate and revol_util are having '%' symbol values and having data type of object. remove % at the end and convert to float

```python
loan_data['int_rate']=loan_data.int_rate.str.strip('%').astype(float)
loan_data['revol_util']=loan_data.revol_util.str.strip('%').astype(float)
```

issue_d, earliest_cr_line are having date values, let convert column data type to date

```python
loan_data['issue_d']=pd.to_datetime(loan_data.issue_d,format='%b-%y')
```

```python
loan_data['earliest_cr_line']=pd.to_datetime(loan_data.earliest_cr_line,format='%b-%y')
```

```python
loan_data.dtypes
```

### Check point 4

We have cleaned the data to the best possible way, we have also fixed the data types of date columns. Now we can create some additional derived columns to enhance our analysis on the loan information
* Assigning numeric values to the category column 'Fully Paid' and 'Charged'
* Derived columns for year and month for issue_dt and earliest credit line

```python
loan_data['code'] = loan_data.loan_status.astype('category').cat.codes
```

```python
loan_data.code.value_counts()
```

```python
loan_data['issue_d_year'] = loan_data.issue_d.dt.year.astype(object)
loan_data['issue_d_month'] = loan_data.issue_d.dt.strftime('%b')
```

```python
loan_data['earliest_cr_line_year'] = loan_data.issue_d.dt.year.astype(object)
loan_data['earliest_cr_line_month'] = loan_data.issue_d.dt.strftime('%b')
```

```python
loan_data['risk_level'] = loan_data.grade.apply(lambda x : 'low' if x in ('A','B') else 'med' if x in ('C','D') else 'high')
```

```python
loan_data.dtypes
```

```python
loan_data.shape
```
